```python
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Masking, LSTM, Dense, Dropout, Flatten, Embedding, concatenate
from tensorflow.keras.utils import plot_model
```

```python
def init_model(size):
    num_input = Input(shape=(size,))
    x = Dense(25)(num_input)
    x = Dropout(0.4)(x)
    for i in range(5):
        x = Dense(25)(x)
        x = Dropout(0.4)(x)
    output = Dense(1)(x)
    model = Model(inputs=num_input, outputs=output)
    return model
base_model = init_model(10)
plot_model(base_model, show_shapes=True)
```
